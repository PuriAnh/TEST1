```python
# bài 1

import torch
import torch.nn as nn

class Softmax(nn.Module):
    def __init__(self):
        super(Softmax, self).__init__()
        # tạo thêm các thành phần khác nếu cần thiết (1)

    def forward(self, x):
        exp_x = torch.exp(x)
        return exp_x / exp_x.sum()

class SoftmaxStable(nn.Module):
    def __init__(self):
        super(SoftmaxStable, self).__init__()
        # (1)

    def forward(self, x):
        # trừ giá trị lớn nhất
        max_x = torch.max(x)
        exp_x = torch.exp(x - max_x)
        return exp_x / exp_x.sum()
```

```python
# bài 3

class Stack:
    def __init__(self, capacity): # tạo Stack với capacity
        self.capacity = capacity
        self.stack = []

    def is_empty(self): # kiểm tra stack rỗng
        return len(self.stack) == 0

    def is_full(self): # kiểm tra stack đầy
        return len(self.stack) == self.capacity

    def pop(self): # loại bỏ phần tử trên cùng và trả về giá trị đó
        if not self.is_empty():
            return self.stack.pop()
        else:
            raise IndexError("Pop from an empty stack")

    def push(self, value): # thêm phần tử vào stack
        if not self.is_full():
            self.stack.append(value)
        else:
            raise OverflowError("Push to a full stack")

    def top(self): # lấy giá trị phần tử trên cùng mà không loại bỏ
        if not self.is_empty():
            return self.stack[-1]
        else:
            raise IndexError("Top from an empty stack")
```

```python
# bài 4
class Queue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue = []

    def is_empty(self): # kiểm tra queue rỗng
        return len(self.queue) == 0

    def is_full(self): # kiểm tra queue đầy
        return len(self.queue) == self.capacity

    def dequeue(self): # loại bỏ phần tử đầu tiên và trả về giá trị đó
        if not self.is_empty():
            return self.queue.pop(0)
        else:
            raise IndexError("Dequeue from an empty queue")

    def enqueue(self, value): # thêm phần tử vào queue
        if not self.is_full():
            self.queue.append(value)
        else:
            raise OverflowError("Enqueue to a full queue")

    def front(self): # lấy giá trị phần tử đầu tiên mà không loại bỏ
        if not self.is_empty():
            return self.queue[0]
        else:
            raise IndexError("Front from an empty queue")
```
